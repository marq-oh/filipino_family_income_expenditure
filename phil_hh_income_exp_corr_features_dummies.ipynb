# Inspiration
Socio-economic classification models in the Philippines has been very problematic. In fact, not one SEC model has been widely accepted. Government bodies uses their own SEC models and private research entities uses their own. We all know that household income is the greatest indicator of one's socio-economic classification that's why the publisher would like to find out the following:

1. Best model in predicting household income
2. Key drivers of household income, we want to make the model as sparse as possible
3. Some exploratory analysis in the data would also be useful

From <https://www.kaggle.com/grosvenpaul/family-income-and-expenditure>

```python
# Function to create 'age-range' column
def setAgeRange(data):
    cat = ''
    if data < 16:
        cat = 'Under 16'
    elif data >= 16 and data <= 24:
        cat = '16 to 24'
    elif data >= 25 and data <= 34:
        cat = '25 to 34'
    elif data >= 35 and data <= 44:
        cat = '35 to 44'
    elif data >= 45 and data <= 54:
        cat = '45 to 54'
    elif data >= 55 and data <= 64:
        cat = '54 to 64'
    elif data >= 65:
        cat = '65 and over'
    else:
        cat = 'Other'
    return cat
```

```python
income_data['Household Head Age Range'] = income_data.apply(lambda x: setAgeRange(x['Household Head Age']), axis=1)
```

## Data Prep

```python
# Remove "anomalies" -> Super Rich and Super Power; Keep only middle-class
income_data_middle = income_data[(income_data['Total Household Income'] >=100000) & (income_data['Total Household Income'] <= 200000)]
```

```python
income_data_middle.head()
```

```python
income_data_middle.shape
```

```python
# Import label encoder 
from sklearn import preprocessing
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column
income_data_middle['Region'] = label_encoder.fit_transform(income_data_middle['Region']) 
income_data_middle['Main Source of Income'] = label_encoder.fit_transform(income_data_middle['Main Source of Income']) 
income_data_middle['Household Head Sex'] = label_encoder.fit_transform(income_data_middle['Household Head Sex']) 
income_data_middle['Household Head Job or Business Indicator'] = label_encoder.fit_transform(income_data_middle['Household Head Job or Business Indicator']) 
income_data_middle['Household Head Age Range'] = label_encoder.fit_transform(income_data_middle['Household Head Age Range']) 
income_data_middle['Household Head Marital Group'] = label_encoder.fit_transform(income_data_middle['Household Head Marital Group']) 
income_data_middle['Level Completed'] = label_encoder.fit_transform(income_data_middle['Level Completed'])
income_data_middle['Household Head Class of Worker Group'] = label_encoder.fit_transform(income_data_middle['Household Head Class of Worker Group'])
# This one was throwing an error -> '<' not supported between instances of 'str' and 'float'
# So added .astype(str) at the end
income_data_middle['Household Head Occupation Group'] = label_encoder.fit_transform(income_data_middle['Household Head Occupation Group'].astype(str))

income_data_middle.head()
```

```python
income_data_middle.shape
```

```python
data_for_model = income_data_middle
```

```python
data_for_model.shape
```

## Finalize model dataset

```python
income_data.info()
```

```python
model_dataset = income_data.iloc[:,[0,1,2,7]]
```

```python
model_dataset = pd.get_dummies(model_dataset, drop_first=False);
model_dataset.head()
```

## New Train and Test Data

```python
X = model_dataset[model_dataset.columns[-27:]]
y = model_dataset['Total Household Income']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

## Train model

```python
# MSJ
model = LinearRegression()
```

```python
# MSJ
model.fit(X_train,y_train)
```

```python
# MSJ: coefficients
# model.coef_,X.columns
coeff_df = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

## Predicting Test Data

```python
# MSJ
predictions = model.predict(X_test)
predictions
```

## Check Model

```python
# MSJ
plt.scatter(y_test,predictions)
```

```python
# MSJ - Evaluate - Check MAE, MSE and RMSE
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# Residuals
sns.distplot((y_test-predictions),bins=50);
```
